```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("../results/results_svd(2).json") as fp:
    metrics = ["precisions", "recalls", "ndcgs", "maps", "rmse", "mae", "rsquared", "exp_var"]
    svd = {k: {metrics[i]: v[i] for i in range(len(v))} for k, v in json.loads(fp.read()).items() }
    
with open("../results/results_vae.json") as fp:
    vae = json.load(fp)    
    
with open("../results/results_knn.json") as fp:
    knn = json.load(fp)
    
```

```python
from matplotlib.transforms import Affine2D

sparsity_levels = list(map(lambda x: float(x), vae))
vae_ndcg = np.array(list(map(lambda x: x['ndcgs'], vae.values())))
svd_ndcg = np.array(list(map(lambda x: x['ndcgs'], svd.values())))
knn_ndcg = np.array(list(map(lambda x: x['ndcg'], knn.values())))


fig, ax = plt.subplots()

trans1 = Affine2D().translate(-0.002, 0.0) + ax.transData
trans2 = Affine2D().translate(0.002, 0.0) + ax.transData
trans3 = Affine2D().translate(0.006, 0.0) + ax.transData
er1 = ax.errorbar(sparsity_levels, vae_ndcg.mean(axis=1), vae_ndcg.std(axis=1), linestyle='None', marker='o', label = "VAE", transform=trans1)
er2 = ax.errorbar(sparsity_levels, svd_ndcg.mean(axis=1), svd_ndcg.std(axis=1), linestyle='None', marker='o', label = "SVD", transform=trans2)
er3 = ax.errorbar(sparsity_levels, knn_ndcg.mean(axis=1), knn_ndcg.std(axis=1), linestyle='None', marker='o', label = "KNN", transform=trans3)

plt.legend()
plt.xticks([0.75, 0.80, 0.85, 0.90, 0.95])
plt.xlabel("Sparsity")
plt.ylabel("NDCG@10")
plt.savefig("../results/sparsity_over_ndcg.png")
plt.show()
```
